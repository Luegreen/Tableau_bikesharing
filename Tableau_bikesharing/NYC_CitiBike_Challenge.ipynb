```python
import pandas as pd
import os 
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
file_to_load = os.path.join("data", "201908-citibike-tripdata.csv")
nyc_bike_df = pd.read_csv(file_to_load)
nyc_bike_df
```

```python
# 2. Check the datatypes of your columns. 
nyc_bike_df.info()
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
#df['dates'] = pd.to_datetime(df['dates'], format='%Y%m%d')
nyc_bike_df["tripduration_long"] = pd.to_datetime(nyc_bike_df['tripduration'], unit='s').dt.strftime("%H:%M:%S")
nyc_bike_df
```

```python
# 4. Check the datatypes of your columns. 
nyc_bike_df.info()
nyc_bike_df["tripduration_long"]
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
nyc_bike_df.to_csv('citibike_data_hms.csv', index = False)
```
